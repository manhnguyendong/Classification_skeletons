# Split model and training with KNN model

```python
#getting dataset have true result
dataset_survey = dataset[dataset['Result'] == 1]
```

```python
dataset_survey
```

```python
dataset_survey.reset_index()
```

```python
dataset_survey.columns
```

```python
X = dataset_survey.iloc[:, 1:-1]
```

```python
X
```

```python
y = dataset_survey.iloc[:, -1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred, y_test) * 100)
```

```python
X = dataset.iloc[:, 1:-1]
y = dataset.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
knn_model = KNeighborsClassifier(n_neighbors=7)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
num_neighbors = [1, 5, 6, 7, 10, 13]
for i in num_neighbors:
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)
    print('Accuracy with n_neighbors %d %f' %(i, accuracy_score(y_test, y_pred)))
```
