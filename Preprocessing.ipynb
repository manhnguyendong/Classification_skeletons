```python
def cvtlabelStringtoNumber(list_type_convert, name_change, dataset):
    for (type, name) in zip(list_type_convert, name_change):
        type_mapping = {x : int(x[-1]) for x in type if (x != 'nan')}
        dataset[name] = dataset[name].map(type_mapping)
        dataset[name] = dataset[name].fillna(0)
        dataset[name] = dataset[name].astype('int')
        print(type_mapping)
```

```python
name_cvt = ['Loai I', 'Loai II', 'Loai III', 'Loai IV', 'Loai V', 'Loai VI', 'Loai VII']
list_type_convert = [loai1, loai2, loai3, loai4, loai5, loai6, loai7]
cvtlabelStringtoNumber(list_type_convert=list_type_convert, name_change=name_cvt, dataset=dataset)
```

```python
map_res = {'có' : 1}
dataset['Result'] = dataset['Result'].map(map_res)
dataset['Result'] = dataset['Result'].fillna(0)
dataset['Result'] = dataset['Result'].astype('int')
```

```python
quality = LabelEncoder()
quality.fit(dataset['CL'])
quality.transform(dataset['CL']) + 1
```

```python
quality.transform(dataset['CL'])
```

```python
quality = list(np.unique(dataset['CL']))
```

```python
dist = {x : (i + 1) for (x, i) in zip(quality, range(1, len(quality) + 1))}
```

```python
dist
```

```python
dataset['CL'] = dataset['CL'].map(dist)
```

# Decode String -> Int
- LoaiI (0 ,I1, I2, I3, ...) -> (1, 2, 3, ...)
- LoaiII (II1, II2, II3, ...) -> (1, 2, 3, ...)
- Loai (R, X) -> (1, 2)
- Result (khong, co) -> (0, 1)
- CL (A, B, C, D ...) -> (1, 2, 3, 4 ...)
